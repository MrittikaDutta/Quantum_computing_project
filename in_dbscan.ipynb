```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
blobs= pd.read_csv('input_data.csv')
```

```python
blobs.head()
```

```python
sns.scatterplot(data=blobs,x='Feature_1',y='Feature_2')
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN()
```

```python
def display_categories(model,data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data,x='Feature_1',y='Feature_2',hue=labels,palette='Set1')
```

```python
model= DBSCAN(eps=2)
display_categories(model,blobs)
```

```python
def display_categories_and_save(model, data, output_file):
    labels = model.fit_predict(data[['Feature_1', 'Feature_2']])
    data['Cluster'] = labels
    sns.scatterplot(data=data, x='Feature_1', y='Feature_2', hue='Cluster', palette='Set1')
    plt.show()
    data.to_csv(output_file, index=False)
```

```python
output_file = 'data_dbscan111.csv'
display_categories_and_save(model, blobs, output_file)
```
