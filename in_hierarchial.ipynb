```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

```python
blobs = pd.read_csv('input_data.csv')
```

```python
print(blobs.head())
```

```python
sns.scatterplot(data=blobs, x='Feature_1', y='Feature_2')
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
def display_categories_and_save(model, data, output_file):
    labels = model.fit_predict(data[['Feature_1', 'Feature_2']])
    data['Cluster'] = labels
    sns.scatterplot(data=data, x='Feature_1', y='Feature_2', hue='Cluster', palette='Set1')
    plt.show()
    data.to_csv(output_file, index=False)
```

```python
output_file = 'data_hierarchical_clustering.csv'
display_categories_and_save(model, blobs, output_file)
```
